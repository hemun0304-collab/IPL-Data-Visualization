```python
# ==========================================
# TASK 3 - IPL DATA VISUALIZATION
# ==========================================

# importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# loading dataset
df = pd.read_csv("C:/Users/SOWMYA N/Downloads/data.csv")

# ------------------------------------------
# checking dataset
# ------------------------------------------

print(df.head())

# ------------------------------------------
# visualization 1 - top winning teams
# ------------------------------------------

plt.figure(figsize=(10,6))

df['winner'].value_counts().head(10).plot(
    kind='bar'
)

plt.title("Top Winning IPL Teams")
plt.xlabel("Teams")
plt.ylabel("Number of Wins")

plt.xticks(rotation=45)

plt.show()

# ------------------------------------------
# visualization 2 - matches per season
# ------------------------------------------

plt.figure(figsize=(10,6))

df['season'].value_counts().sort_index().plot(
    kind='line',
    marker='o'
)

plt.title("IPL Matches Per Season")
plt.xlabel("Season")
plt.ylabel("Number of Matches")

plt.show()

# ------------------------------------------
# visualization 3 - toss decision analysis
# ------------------------------------------

plt.figure(figsize=(7,5))

sns.countplot(
    x='toss_decision',
    data=df
)

plt.title("Toss Decision Analysis")
plt.xlabel("Decision")
plt.ylabel("Count")

plt.show()

# ------------------------------------------
# visualization 4 - top player of the match winners
# ------------------------------------------

plt.figure(figsize=(12,6))

df['player_of_match'].value_counts().head(10).plot(
    kind='bar'
)

plt.title("Top Player of the Match Winners")
plt.xlabel("Players")
plt.ylabel("Awards")

plt.xticks(rotation=45)

plt.show()

# ------------------------------------------
# visualization 5 - city wise matches
# ------------------------------------------

plt.figure(figsize=(12,6))

df['city'].value_counts().head(10).plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Top Cities Hosting IPL Matches")

plt.ylabel("")

plt.show()

# ------------------------------------------
# visualization 6 - heatmap
# ------------------------------------------

plt.figure(figsize=(6,4))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()

# ------------------------------------------
# final insights
# ------------------------------------------

print("Mumbai Indians and CSK are among the top winning teams.")
print("IPL matches increased over seasons.")
print("Most teams prefer fielding after winning toss.")
print("Some players consistently win player of the match awards.")
```
